```python
x=[1,2,3,4,5]
sales=[10,20,30,40,50]
profit=[2,3,4,7,8]

plt.plot(x,sales,label='sales',color='blue')
plt.plot(x,profit,label='profit',color='green')
plt.xlabel('Day')
plt.ylabel('Amount')
plt.title("sales vs profit")
plt.legend()
plt.grid(True)
plt.show()
```

```python
products=["Laptop", "Mouse", "Keyboard", "Monitor", "Printer"]
sales=[100,250,150,75,50]
plt.bar(products,sales,label='sales',color='pink')
plt.title('product sales')
plt.xlabel('product')
plt.ylabel('Unit sold')
plt.grid(axis='y')
plt.show()
```

```python
plt.barh(products, sales, color='lightgreen',edgecolor='orange')
plt.title("Product Sales (Horizontal)")
plt.xlabel("Units Sold")
plt.ylabel("Product")
plt.grid()
plt.show()
```

```python
ages = [22, 25, 30, 22, 35, 40, 26, 30, 27, 30, 25, 26, 24, 28, 29, 32]

plt.hist(ages, bins=5, color='orange', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
toys = ['Laptop', 'Mouse', 'Keyboard', 'Monitor']
sales = [300, 100, 150, 50]

plt.pie(sales, labels=toys, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'yellow'],startangle=90)
plt.title("Sales Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Rent', 'Food', 'Transport', 'Internet', 'Savings']
amounts = [4000, 2500, 1000, 500, 2000]

# Plot
plt.pie(amounts, labels=categories, autopct='%1.1f%%',
        colors=['lightblue', 'lightgreen', 'orange', 'pink', 'violet'],
        startangle=90)

plt.title("Monthly Budget Allocation")
  # Make circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
height = [150, 155, 160, 165, 170, 175, 180]
weight = [45, 50, 53, 58, 65, 68, 75]

# Scatter plot
plt.scatter(height, weight, color='blue', marker='^', s=100)

# Customizations
plt.title("Height vs Weight of Students")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.grid(True)

plt.figure(figsize=(8, 8))

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 25, 15]
y2 = [5, 15, 10, 20, 30]

# 1 row, 2 columns
plt.subplot(1, 2, 1)
plt.plot(x, y1, color='blue')
plt.title("Sales")

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='green')
plt.title("Profit")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 2))  # Width x Height in inches

plt.plot(x, y1, color='red')
plt.title("Figure with Custom Size")
plt.show()
```

```python
plt.plot(x, y1, color='purple', linestyle='--', marker='o')
plt.title("Styled Line Plot")
plt.grid(True)
plt.show()
```
